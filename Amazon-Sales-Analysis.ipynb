### Size

```python
ex = sns.countplot(x='Size', data=df)
```

#### Note: From the above graph we observed that most ofthe people buys M-Size clothing

## Group By

#### The group by function in pandas is used to group data based on one or more columns in a DataFrame

```python
df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty', ascending=False)
```

```python
S_Qty = df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty', ascending=False)
sns.barplot(x='Size', y='Qty', data=S_Qty)
```

#### Note: From above graph we can see that most of the Qty buys M-Size in the sales

```python
#histogram
df['Size'].hist()
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10, 5))
plt.hist(column_data, bins=20, edgecolor='RED')
#plt.xticks(rotation=90)
plt.show()
```

#### Note: From the above Graph we observed that most of the buyers are buying T-shirt

```python
#Checking B2B Data by using pie chart
B2B_Check = df['B2B'].value_counts()

#Plot the pie chart
plt.pie(B2B_Check, labels=B2B_Check, autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

```python
#Checking B2B Data by using pie chart
B2B_Check = df['B2B'].value_counts()

#Plot the pie chart
plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

#### Note: From the above chart we can see that maximum i.e. 99.2% of buyers are B2C and 0.8% are B2B buyers

```python
#Prepare data for scatter plot
x_data = df['Category']
y_data = df['Size']

#Plot the scatter plot
plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
#PLot count of cities by State
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='ship-state', hue='ship-state')
plt.xlabel('ship-state')
plt.ylabel('Count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
#top_10_states
top_10_states = df['ship-state'].value_counts().head(10)
#PLot count of cities by State
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['ship-state'].isin(top_10_states.index)], x='ship-state', hue='ship-state')
plt.xlabel('ship-state')
plt.ylabel('Count')
plt.title('Distribution of State')
plt.xticks(rotation=45)
plt.show()
```

#### Note: From the above graph we can observe the most of buyers are from Maharashtra
